```python
tweets = []

def tweets_to_csv(username, count): 
    for tweet in api.user_timeline(id=username, count=count):
        
        # Adding to list that contains all tweets
        tweets.append((tweet.created_at,tweet.id,tweet.text))
        
        # Creation of dataframe from tweets list
        tweetsdf = pd.DataFrame(tweets,columns=['Datetime', 'Tweet Id', 'Text'])
        
        # Converting dataframe to CSV
        tweetsdf.to_csv('{}-tweets.csv'.format(username)) 
```

```python
username = 'aeriebeaar'
count = 300

# Calling function to turn username's past X amount of tweets into a CSV file
tweets_to_csv(username, count)
```

```python
tweetsdf['RT'] = df['Text'].str[:2]
tweetsdf
```

```python
for i in range(len(tweetsdf['Text'])):
    if tweetsdf['Text'][i][0] == '@':
        tweetsdf['TweetType'][i] = 'Mention'
    elif tweetsdf['RT'][i] == 'RT':
        tweetsdf['TweetType'][i] = 'RT'
    else:
        tweetsdf['TweetType'][i] = 'Tweet'
        
```

```python
tweetsdf
```

```python
tweetsdf['TweetType'].value_counts().plot(kind='bar');
plt.xlabel("Tweet Type", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Count of Tweets by Type", y=1.02);
```
