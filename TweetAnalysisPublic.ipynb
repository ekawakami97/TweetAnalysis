```python
tweets = []

def tweets_to_csv(username, count): 
    for tweet in api.user_timeline(id=username, count=count):
        
        # Adding to list that contains all tweets
        tweets.append((tweet.created_at,tweet.id,tweet.text))
        
        # Creation of dataframe from tweets list
        tweetsdf = pd.DataFrame(tweets,columns=['Datetime', 'Tweet Id', 'Text'])
        
        # Converting dataframe to CSV
        tweetsdf.to_csv('{}-tweets.csv'.format(username)) 
```

```python
username = 'aeriebeaar'
count = 300

# Calling function to turn username's past X amount of tweets into a CSV file
tweets_to_csv(username, count)
```

## 3. Data manipulation/ Exploratory Analysis

Once we have the dataframe ready to go, I'm going to clean the data so I can use it for some analysis. I first see what the datatypes are. 

Then, I add a date column since we don't need an exact time for our tweets. 

Afterwards, I categorize each tweets into Mentions, RT, and Tweet using their indexes.



```python
#datatypes
tweetsdf.dtypes 

#object means string 
```

```python
# make date column
tweetsdf['Date'] = df['Datetime'].str[:10]
tweetsdf
```

```python
#tweetsdf.drop(columns = ['year'])
```

```python
tweetsdf['TweetType'] = ''
```

```python
tweetsdf['RT'] = df['Text'].str[:2]
tweetsdf
```

```python
for i in range(len(tweetsdf['Text'])):
    if tweetsdf['Text'][i][0] == '@':
        tweetsdf['TweetType'][i] = 'Mention'
    elif tweetsdf['RT'][i] == 'RT':
        tweetsdf['TweetType'][i] = 'RT'
    else:
        tweetsdf['TweetType'][i] = 'Tweet'
```

```python
tweetsdf
```

Now we get into the visualizations. 

I start with just seeing how many of each category of tweets I engage with the most. It seems that I retweet most of the content on my Twitter. 

```python
tweetsdf['TweetType'].value_counts().plot(kind='bar');
plt.xlabel("Tweet Type", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Count of Tweets by Type", y=1.02);
```

I also want to create a word cloud to see out of what I have tweeted, what word or topics (in the last couple of months) I've tweeted about the most. 

I start with importing the necessary packages for a word cloud.

Then, I make the Text column into one long string to make it into the format necessary for the Wordcloud package. 

```python
# Import packages 
from wordcloud import WordCloud, STOPWORDS 
```

```python
# extract the text
text = tweetsdf.Text[0]
```

Next I see how many words are in the combined string of tweets.

```python
text = " ".join(review for review in tweetsdf.Text)
print ("There are {} words in the combination of all review.".format(len(text)))
```

I create my first word cloud

```python
# Generate a word cloud image
wordcloud = WordCloud(background_color="white").generate(text)



# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure(1, figsize=(12, 12))
plt.show()
```

It seems like in the text, *RT* and *https* are all seen as its own words. This doesn't give us any context of what I've been tweeting about. Also, there are individual letters like *t* and *s* that are on it's own. This is probably from contraction words like can't or I'm. 

So we proceed to clean the string - get rid of apostraphes. 

Then we add stop words for the word cloud so they don't count those as valid words to count

```python
# create stop words and get rid of punctuations 
text = text.replace("‚Äô", "")
text

stopwords = set(STOPWORDS)
stopwords.update(["https", "RT", "t", "co"])
```

```python
from matplotlib.pyplot import figure #make figure larger

wordcloud = WordCloud(max_words=200,
                      stopwords=stopwords, background_color="white",
                     scale = 3).generate(text)


fig = plt.figure(1, figsize=(10, 10))
plt.axis("off")
plt.imshow(wordcloud)
plt.show()
```

This seems like a more valid word cloud.

In the past couple of months, It seems I talk about myself, my needs, my wants, and things related to voting. It seems fitting for the events that have occured in the past couple of months.
